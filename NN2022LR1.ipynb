# **Numpy**

```python
x1 =  np.array([1,2,3])
print()
x2 =  np.array([4,5,6])
print()

x_dot = np.dot(x1,x2)
print("Скалярний добуток = ", x_dot)

x_cross = np.cross(x1,x2)
print("Векторний добуток = ", x_cross)
```

```python
m1 = np.array([[1,2,3],[4,5,6]])
m2 = np.array([[6,5],[4,3],[2,1]])

m = np.dot(m1,m2)
print("Матричний добуток = ", m)
```

```python
print("Розмірність масиву = ",np.shape(m1))
```

```python
print("Old array = ", x1)
new_x = np.reshape(x1,(3,1))
print("New array = ", new_x)
```

```python
x = np.arange(5)
print("Arange function = ", x)
x = np.random.random()
print("Random function = ", x)
```

```python
print("Array operations")
x = np.arange(5)+1
y = np.flip(x)
print(x)
print(y)
print("Sum = ", x+y)
print("Diff = ", x-y)
print("Prod = ", x*y)
print("Div = ", x/y)
```

# **Matplotlib**

```python

x = np.array([-2,-1,0,1,2])
y_hat = x**2
y = (x**2) + np.random.random()
df = pd.DataFrame(np.ones((5,5)), columns=["x","y","y_hat","mAE","mSE"])
df["x"] = x
df["y"] = y
df["y_hat"] = y_hat
df["mAE"] = df["y"] - df["y_hat"]
df["mSE"] = (df["y"] - df["y_hat"])**2
df.to_csv("lr1.csv")
files.download("lr1.csv")
plt.plot(x,y, color = "red")
plt.show()
print(y)
print(y_hat)
df.head()
```
