```python
import numpy as np
from sklearn import datasets
from sklearn import metrics 
from sklearn.linear_model import LogisticRegression
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
import matplotlib.pyplot as plt 
from sklearn.metrics import confusion_matrix
```

```python
x, y = datasets.make_classification(n_samples=1000,n_informative=1, n_features=2, n_classes=2, n_redundant=0, n_clusters_per_class=1,class_sep = 2) 
fig, ax = plt.subplots(figsize=(9,6))
plt.xlabel('x', fontsize = 16)
plt.ylabel('y', fontsize = 16)
plt.scatter(x[:,0], x[:,1] ,c=y)
```

```python
min1, max1 = x[:, 0].min()-1, x[:, 0].max()+1
min2, max2 = x[:, 1].min()-1, x[:, 1].max()+1
arr_x1 = np.arange(min1, max1, 0.01)
arr_x2 = np.arange(min2, max2, 0.01)
X, Y = np.meshgrid(arr_x1, arr_x2)
r1, r2 = X.flatten(), Y.flatten()
r1, r2 = r1.reshape((len(r1), 1)), r2.reshape((len(r2), 1))
grid = np.hstack((r1,r2))

model = LinearDiscriminantAnalysis()
model.fit(x, y)
pred_y = model.predict(grid)
Z = pred_y.reshape(X.shape)
plt.contourf(X,Y,Z,cmap='Pastel2')
plt.scatter(x[:,0], x[:,1], s=50, c=y)
plt.show()
```

```python
t_n, f_p, f_n, t_p = confusion_matrix(y, model.predict(x)).ravel()
(t_n, f_p, f_n, t_p)
```
